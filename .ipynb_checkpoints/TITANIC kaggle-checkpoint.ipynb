```python
from keras.layers import Dense, Dropout
from keras.models import Sequential 

from keras.utils import np_utils

from matplotlib import pyplot as plt
```

```python
import pandas as pd
```

```python
train = pd.read_csv('./titanic/train.csv')
test = pd.read_csv('titanic/test.csv', sep=',')
gender_submission = pd.read_csv('titanic/gender_submission.csv', sep=',')
print(test)
```

```python
X_train = train[['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked']].copy()
Y_train = train[['Survived']].copy()

#X_test = test[['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked']].copy()
#Y_test = test[['Survived']].copy()

Y_train = np_utils.to_categorical(Y_train, 2)
#Y_test = np_utils.to_categorical(Y_test, 2)
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(2, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python


 model.fit(X_train, Y_train, batch_size=128, epochs=10, validation_split=0.2, verbose=1)
```
